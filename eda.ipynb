```python
import pandas
import matplotlib.pyplot as plt
plt.style.use("seaborn-darkgrid")
```

TSV file generated with command
`find cleaned -name "*.txt" | xargs python3 parser.py > compiled.tsv`

```python
df = pandas.read_csv("compiled.tsv",sep="\t")
df.info()
```

```python
plt.title("Histogram of Scores")
df["Score"].hist(bins=[0,1001,5001,10001,50001])
```

```python
df["Length"]=df["Quote"].str.len()
plt.title("Score vs. Quote Length")
plt.xlabel("Length in Chars")
plt.ylabel("Score")
plt.plot(df["Length"],df["Score"],'r.',alpha=0.15)
```

```python
df.sort_values(by="Score",ascending=False)
```
